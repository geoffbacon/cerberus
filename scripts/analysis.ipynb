```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
def plot_performance(task):
    plt.figure(figsize=(10,8))
    results = pd.read_csv("results.csv")
    sns.scatterplot(x="num", y="score", data=results[results["task"] == task], s=50);
    if task == "pos-tagging":
        title = "POS tagging"
    elif task == "classification":
        title = "Subject number classification"
    fs = 20
    plt.title(title, fontsize=fs)
    plt.xlabel("Number of training examples", fontsize=fs)
    plt.ylabel("Accuracy", fontsize=fs)
    plt.ylim((0, 100))
    plt.xlim((0, 3000))
```

```python
plot_performance("pos-tagging")
```
